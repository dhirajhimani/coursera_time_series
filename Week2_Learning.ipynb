```r
x=NULL
x[1]=0
for (i in 2:1000) {
    x[i] = x[i-1] + rnorm(1)
}
#print(x)
plot(x, main="A random walk Non TS")
random_walk_ts = ts(x)
plot(random_walk_ts, main="A random walk  TS")
(acf(random_walk_ts))
```

```r
random_walk_diff = diff(random_walk_ts)
print(random_walk_diff)
```

```r
plot(random_walk_diff, main="Difference gives white noise")
acf(random_walk_diff)
```

```r
#MAP at t
t=3
q=t+1 # For ease of calculation

# Generate noise
noise=rnorm(10000)

# Introduce a variable
ma_2=NULL

# Loop for generating MA(2) process

for(i in q:10000) {
    sub_noise = noise[i]
    
    for(j in 1:t) {
        #print(j)
        sub_noise = sub_noise + (.8 - .1*q)*noise[i-j]
    }
    ma_2[i]=sub_noise
    #print(i)
    #print( ma_2[i])
}

# Shift data to left by 2 units
moving_average_process=ma_2[q:10000]

# Put time series structure on a vanilla data
moving_average_process=ts(moving_average_process)

# Partition output graphics as a multi frame of 2 rows and 1 column
par(mfrow=c(2,1))

# plot the process and plot its ACF
plot(moving_average_process, main='A moving average process of order 2', ylab=' ', col='blue')
acf(moving_average_process, main='Correlogram of a moving average process of order 2')
```

```r
# Simulating a non-stationary time series

# Set seed so thet we generate the same dataset
set.seed(2017)
# time variable 
t=seq(0,1,1/100)
# generate a time series
some.time.series=2+3*t+ rnorm(length(t))

# obtain acv for this time series below
(acf(some.time.series, type='covariance'))
```

```r
# Simulating MA(4) process.
# X_t= Z_t+0.2 Z_(t-1)+0.3 Z_(t-2)+ 0.4 Z_(t-3)

set.seed(2^10)
z=NULL
z=rnorm(1000)
data=NULL
for(i in 4:1000){
  data[i-3]=z[i]+0.2*z[i-1]+0.3*z[i-2]+0.4*z[i-3]
  }
data=ts(data)

# find acf below
# Partition output graphics as a multi frame of 2 rows and 1 column
par(mfrow=c(2,1))

# plot the process and plot its ACF
plot(data, main='A moving average process of order 4', ylab=' ', col='blue')
(acf(data, main='Correlogram of a moving average process of order 4'))

```
